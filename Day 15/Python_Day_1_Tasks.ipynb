3. Write a program to do the following operations :
- Read any two positive integer operands (say op1 & op2) and one character type operator (say opr). Note that opr is any mathematical operator.
- Depending upon the operator, do the appropriate operation. e. g. if opr is ‘+’ then the display the value obtained by evaluating the expression (op1 + op2).

```python
op1 = int(input("Enter the first number: "))
print("The first number is", op1)
op2 = int(input("Enter the second number: "))
print("The second number is", op2)
opr = input("Enter the operator (+, -, *, /): ")
print("The operator is", opr)
if opr == '+':
    print("The sum is", op1 + op2)
elif opr == '-':
    print("The difference is", op1 - op2)
elif opr == '*':
    print("The product is", op1 * op2)
elif opr == '/':
    if op2 != 0:
        print("The quotient is", op1 / op2)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operator. Please use +, -, *, or /.")
```

 4. An electronic component vendor supplies three products : transistors,resistors and capacitors. The vendor gives a discount of 10% on orders for transistors if the order is for more than Rs. 1000. On orders of more than Rs. 100 for resistors, a discount of 5% is given, and a discount of 10% is given on orders for capacitors of value more than Rs. 500. Assume that the numeric codes 1,2 and 3 are used for transistors, capacitors and resistors respectively. Write a program that reads the product code and the order amount and prints out the net amount that the customer is required to pay after discount.

```python
product_code = int(input("Enter product code (1=transistor, 2=capacitor, 3=resistor): "))
amount = float(input("Enter order amount: "))

discount = 0.0
if product_code == 1:
    if amount > 1000:
        discount = 0.10
elif product_code == 2:
    if amount > 500:
        discount = 0.10
elif product_code == 3:
    if amount > 100:
        discount = 0.05
else:
    print("Invalid product code.")

if product_code in (1, 2, 3):
    net_amount = amount * (1 - discount)
    print(f"Net amount after discount: {net_amount:.2f}")
```

10. Write a program that reads three positive numbers a, b, c and determines whether they can form the three sides of a triangle. If yes, determine whether the triangle will be an obtuse-angle, or a right-angle or an acuteangle triangle. If the triangle is an acute angle triangle, determine further whether the triangle is equilateral, isosceles or scalene.

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

sides = sorted([a, b, c])
a, b, c = sides

if a <= 0 or b <= 0 or c <= 0:
    print("Sides must be positive numbers.")
elif a + b <= c:
    print("The numbers cannot form a triangle.")
else:
    if abs(c**2 - (a**2 + b**2)) < 1e-9:
        triangle_type = "right-angle"
    elif c**2 > a**2 + b**2:
        triangle_type = "obtuse-angle"
    else:
        triangle_type = "acute-angle"

    print("The triangle is a", triangle_type, "triangle.")

    if triangle_type == "acute-angle":
        if abs(a - b) < 1e-9 and abs(b - c) < 1e-9:
            shape = "equilateral"
        elif abs(a - b) < 1e-9 or abs(b - c) < 1e-9 or abs(a - c) < 1e-9:
            shape = "isosceles"
        else:
            shape = "scalene"
        print("It is also an", shape, "triangle.")
```

11. A function f is defined as follows:

f(x) = 

-     = ax^3 – bx^2 + cx – d, if x > k

-     = 0, if x = k

-     = -ax^3 + bx^2 - cx + d, if x < k 

Write a program that reads a, b, c, d, k and x and prints the value of f(x).

```python
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
d = float(input("Enter d: "))
k = float(input("Enter k: "))
x = float(input("Enter x: "))

if x > k:
    fx = a * x**3 - b * x**2 + c * x - d
elif x == k:
    fx = 0
else:
    fx = -a * x**3 + b * x**2 - c * x + d

print("f(x) =", fx)
```

14. SUM = 1 + 4 – 9 + 16 – 25 + 36 - … and so on

```python
n = int(input("Enter the number of terms: "))
if n > 0:
    total = 0
    for i in range(1, n + 1):
        square = i * i
        sign = 1 if i == 1 or i % 2 == 0 else -1
        total += sign * square
    print("SUM =", total)
else:
    print("Please enter a positive integer for the number of terms.")
```

16. SUM = x – x^3/3! + x^5/5! - x^7/7! + x^9/9! -… and so on

```python
import math
x = float(input("Enter x: "))
terms = int(input("Enter the number of terms: "))
if terms > 0:
    total = 0.0
    for i in range(terms):
        exponent = 2 * i + 1
        term = ((-1) ** i) * x**exponent / math.factorial(exponent)
        total += term
    print("SUM =", total)
else:
    print("Please enter a positive integer for the number of terms.")
```

34. To find sum of all odd digits and even digits from a given positive number.

```python
number = int(input("Enter a positive number: "))
if number >= 0:
    odd_sum = 0
    even_sum = 0
    for digit in str(number):
        if digit.isdigit():
            value = int(digit)
            if value % 2 == 0:
                even_sum += value
            else:
                odd_sum += value
    print("Sum of odd digits:", odd_sum)
    print("Sum of even digits:", even_sum)
else:
    print("Please enter a non-negative number.")
```

35. To find sum of all prime digits from a given number.

```python
number = input("Enter a number: ")
prime_digits = {2, 3, 5, 7}
prime_sum = 0
for ch in number:
    if ch.isdigit() and int(ch) in prime_digits:
        prime_sum += int(ch)
print("Sum of prime digits:", prime_sum)
```
